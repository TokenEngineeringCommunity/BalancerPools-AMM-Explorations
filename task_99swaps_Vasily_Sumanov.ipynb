```python
from decimal import Decimal
import pprint
pp = pprint.PrettyPrinter(indent=4)
#-----------------
def calc_in_given_out(
            token_balance_out: Decimal,#B0
            token_balance_in: Decimal,#Bi
            token_amount_out: Decimal,#A0
            token_weight_in: Decimal,#
            token_weight_out: Decimal,#
            swap_fee: Decimal):
     
        token_amount_in=token_balance_in*(token_balance_out/(token_balance_out-token_amount_out)-1)/(1-swap_fee)

        return token_amount_in #Ai
#-----------------------------
def invariant(pool, weight):
    return pool[0]**weight[0]*pool[1]**weight[1]
```

```python
#Swap 1 time - by 99 token B with a fee of 0.3%
pool=[100,100]
weight = [0.5,0.5]
fee=0.003
amount_out = 99
#------------
print('Initial Invariant = ', invariant(pool, weight))
swap= calc_in_given_out(pool[1], pool[0],amount_out, weight[0],weight[1], fee)
print('calculated amount of token_in = ', swap)
pool[0]=pool[0]+swap
pool[1]=pool[1]-amount_out
commision = swap*fee
print('Final Balances: Token A= ', pool[0],', Token B= ', pool[1],', Commision =', commision, ', Final Invariant = ', invariant(pool, weight))
```

```python
#Swap 99 time - by 1 token B with commision fee =0,3%
pool1=[100,100]
weight = [0.5,0.5]
# let's compute the result for fee=0.3%
fee=0.003
amount_out = 1
data1=[[0,100,100, 0, 100]]#step, Token 1, Token 2, Total Commision, invariant
commision=0
token_a = 0
comm=[]
print('Initial Invariant = ', invariant(pool1, weight))
for i in range (1,100):
    swap=calc_in_given_out(pool1[1], pool1[0],amount_out, weight[0],weight[1],  fee)
    print('Token_in amount = ', swap)
    pool1[0]=pool1[0]+swap
    pool1[1]=pool1[1]-amount_out
    commision = commision + swap*fee
    comm.append(commision)
    token_a=token_a+swap
    inv= invariant(pool1, weight)
    data1.append([i, pool1[0], pool1[1], commision, inv])
print('Token A= ', pool1[0],', Token B= ', pool1[1], ', Total commision= ', commision, ', Final Invariant = ', invariant(pool1, weight))
print('--------------------------------------------')
pp.pprint(data1)
print(token_a)

```

```python
#Plot chart of commision 
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
x=np.arange(0,100,1)
fig = make_subplots(subplot_titles=('Commision for Swap 99 times' ))
fig.add_trace(go.Scatter(x=x, y=comm, line=dict(color='blue'), name="Commision"), secondary_y=False)
#fig.add_trace(go.Scatter(x=x1, y=z1, line=dict(color='green'), name="Token 2"), secondary_y=False)


fig.update_layout(title_text='<b>Swap fee accumulated in Token A</b>')
fig.update_xaxes(title_text='Swap number')

fig.show()
```

```python
# let's compute the result for ZERO fee
pool2=[100,100]
weight = [0.5,0.5]
print('Start invariant = ', invariant(pool2, weight))
fee=0
amount_out = 1
data2=[[0,100,100,0,100]]#step, Token 1, Token 2, Total Commision, invariant
for i in range (1,100):
    swap=calc_in_given_out(pool2[1], pool2[0],amount_out, weight[0],weight[1], fee)
    print('Token in amount = ', swap)
    pool2[0]=pool2[0]+swap
    pool2[1]=pool2[1]-amount_out
    inv= invariant(pool2, weight)
    data2.append([i, pool2[0], pool2[1],0, inv])
print('Token A= ', pool2[0],' Token B= ', pool2[1], ', Commision = 0', ' , Final Invariant = ', invariant(pool2, weight))
```
